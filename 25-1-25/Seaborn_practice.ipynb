```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

## Visulazie Distrubutions
- Histogram
- KDE plot(kernal density estimation)

```python
sns.histplot(data=tips,x='total_bill',kde=True,bins=20,color='grey')
plt.title('Distribution of total bills')
plt.show()
```

```python
sns.kdeplot(data=tips,x='total_bill')
plt.show()
```

## Categorical plots
- Bar plot
- Box plot

```python
sns.barplot(data=tips,x='day',y='total_bill',palette=['slategrey','mistyrose','lightgreen'])
plt.title('Average Total bill by day')
plt.show()
```

```python
# tips.groupby('day')['total_bill'].mean()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
plt.title('Box plot of total bills by day and Gender')
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time')
plt.title('SCatter plot of Total bill vs Tip')
plt.show()
```

## relation  ships
- lineplot
- pairplot
- heatmap


```python
sns.lineplot(data=tips,x='size',y='tip',marker='o')
plt.title('Line plot of party Size vs tip')
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex')
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True)
plt.title('Corelation matrix')
plt.show()
```
