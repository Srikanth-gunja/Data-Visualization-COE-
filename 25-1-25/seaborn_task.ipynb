```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
df
```

```python
sns.barplot(data=df,x='Pclass',y='Survived',palette=['slategrey','mistyrose','lightgreen'])
plt.title('Survival rates across different classes')
plt.xlabel('Passeneger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.scatterplot(data=df,x='Age',y='Fare',hue='Survived')
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.pairplot(df,vars=['Age','Fare','Pclass'],hue='Survived')
plt.show()
```

```python
sns.heatmap(df.iloc[:,1:].corr(numeric_only=True),annot=True) # Exclude Passenger Id
plt.title('Correlation ')
plt.show()
```
